```python
import gensim.downloader
from gensim.models import KeyedVectors

word2vec_vectors = gensim.downloader.load('word2vec-google-news-300')

word2vec_vectors["cat"]
```

```python
word2vec_vectors["fish"]
```

```python
import numpy
from numpy import dot
from numpy.linalg import norm

cat_vec = word2vec_vectors["cat"]
fish_vec = word2vec_vectors["fish"]

# 코사인 유사도 계산
cos_sim = dot(cat_vec, fish_vec) / (norm(cat_vec) * norm(fish_vec))

print(cos_sim)
```
